```python
import re
import generic_io
import numpy as np
from gensim.models import Word2Vec
import json_lines
from nltk.tokenize import TweetTokenizer
```

```python
data = generic_io.load_from_file('veri/tweets/Italy-earthquake-day1-tweets1.jsonl', file_format='jsonl')
```

```python
data_ = data[:100]
```

```python
tknzr = TweetTokenizer()
#s0 = "This is a cooool #dummysmiley: :-) :-P <3 and some arrows < > -> <--"
#tknzr.tokenize(s0)
```

```python
tweetlist = []
            
for item in data:
    if re.search("https.*$", item['text']):
        item['text'] = re.sub("https.*$", "", item['text'])
    if re.search("@.* ", item['text']):
        item['text'] = re.sub("@.* ", "", item['text'])
    tokenized_tweet = tknzr.tokenize(item['text'])
    tweetlist.append(tokenized_tweet)
```
